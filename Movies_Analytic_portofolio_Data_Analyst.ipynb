# Machine Learning Workflow in Movie Analytics

```python
new_df.select_dtypes(include= 'number')
##menfilter data2 yang numerical
```

```python
def select_columns(df:pd.DataFrame, feature: list, target: list='rating'): #ratingnya parameter = y

    x = df[feature]
    y = (df[target] >=3).astype('int')

    return x,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature =['budget','popularity','vote_average','vote_count','runtime']
#y sudah di masukin ke parameter yang ada diatas sehingga tidak perluh di masukan lagi
x, y = select_columns(new_df, feature = feature)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3, random_state=42)#random state digunakan agar saat di run di komputer lain menghasilkan data/hasil yang sama
```

```python
x_train #cara menghitung data x_test dan train sudah benar atau tidak saat di split dengan cara (xtest/(xtest+xtrain))
```

```python
310/(310+722 )
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax =MinMaxScaler()
```

```python
minmax.fit(x_train)
```

```python
x_train_scaled= minmax.transform(x_train)
x_test_scaled= minmax.transform(x_test)
```

```python
x_train_scaled #aray data
```

```python
pd.DataFrame(x_train_scaled, columns= minmax.get_feature_names_out()) #merubah aray data menjadi frame data kembali
```

```python
def scaling(df, scaler): # fungsion digunakan bisa berulang-ulang agar tidak membuat program lagi
  df['runtime'] = df['runtime'].div(60)
  Scaled_aray = scaler.transform(df)
  scaled_df = pd.DataFrame(Scaled_aray, columns= scaler.get_feature_names_out())

  return scaled_df
```

```python
scaling(x_train,minmax)
```

```python
x_train_scaled = scaling(x_train, minmax)
x_test_scaled = scaling(x_test,minmax)
```

```python
x_test_scaled
```

```python
knn_best.predict(x_test)
```

```python
recall_ =[]
precision_ = []
f1_score_ =[]
accuracy_ =[]
for i in range(1, 10) :
  y_best_knn_proba =knn_best.predict_proba(x_test)[:, 1]
  recall_.append(recall_score(y_test,y_best_knn_proba >= i/10))
  precision_.append(precision_score(y_test,y_best_knn_proba >= i/10))
  f1_score_.append(f1_score(y_test,y_best_knn_proba >= i/10))
  accuracy_.append(accuracy_score(y_test,y_best_knn_proba >= i/10))
```

```python
pd.DataFrame({'recall': recall_,
              'precison' : precision_,
              'f1_score' : f1_score_,
              'accuracy' : accuracy_})
```

```python
metrics =pd.DataFrame({'recall': recall_,
              'precison' : precision_,
              'f1_score' : f1_score_,
              'accuracy' : accuracy_})
```

```python
metrics['fpr']  = 1- metrics['recall'] #force positive rate
```

```python
metrics.plot(x='fpr', y='recall')
```
